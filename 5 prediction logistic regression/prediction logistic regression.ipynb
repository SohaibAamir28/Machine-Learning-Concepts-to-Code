````python
from sklearn.linear_model import LogisticRegression

# Step 1: Dataset
X = [[40], [50], [60], [70], [80], [90]]
y = [0, 0, 1, 1, 2, 2]  # 0=Fail, 1=Pass, 2=Distinction

# Step 2: Create model
model = LogisticRegression(
    multi_class='multinomial',
    solver='lbfgs'
)

# Step 3: Train model
model.fit(X, y)

# Step 4: Predict result for 65 marks
print("Predicted Class:")
print(model.predict([[65]]))

# Step 5: Predict probabilities
print("\nPredicted Probabilities:")
print(model.predict_proba([[65]]))
````

**üü¢ WHY DO WE NEED POLYNOMIAL LOGISTIC REGRESSION?**

**First, recall NORMAL Logistic Regression:**

Draws a straight line to separate classes

Works only when data is linearly separable

‚ùå Problem:

Many real-world problems are curved, circular, or non-linear

Example:

Exam marks vs result

Age vs disease risk

Experience vs promotion

**üëâ A straight line is not enough**

**üß† IDEA BEHIND POLYNOMIAL LOGISTIC REGRESSION**

>‚ÄúDon‚Äôt change the algorithm ‚Äî change the FEATURES.‚Äù


We create new features like:

      3                2
    x                x

This allows logistic regression to:

**‚úî Draw curved decision boundaries**

We will use a simple real-life example again:

üéØ Example:

Marks ‚Üí Result

But now the relationship is non-linear

**üîπ STEP 1: Import Required Tools**

üëâ This is the key difference from normal logistic regression

````python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures

# Step 1: Dataset
X = [[20], [30], [40], [50], [60], [70], [80]]
y = [0, 0, 0, 1, 1, 1, 0]  # 0=Fail, 1=Pass

# Step 2: Polynomial feature generator
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Step 3: Create model
model = LogisticRegression()

# Step 4: Train model
model.fit(X_poly, y)

# Step 5: Test with new value
test_mark = [[55]]
test_poly = poly.transform(test_mark)

print("Predicted Class:")
print(model.predict(test_poly))

print("\nPredicted Probabilities:")
print(model.predict_proba(test_poly))
````

```
0 = Fail
1 = Pass
```

üëâ Prediction: PASS



**Predicted Probabilities:**

| 1     | Pass    |         |

üëâ The model is confident that the student passes
